```python
# a name given to a variable, constant, function, class, object etc.

# Variable - a
# a = 10

# Constant - MAX_VAL
# MAX_VAL = 100

# function - add
# def add(a,b):
#     return a+b

# class - MyClass
# class MyClass:
#     def __init__(self, a,b):
#         self.a = a
#         self.b = b

# Object - obj
# obj = MyClass(10,20)

# Set of rules to define identifier
#     1. Allowed characters - A to Z, a to z, 0 to 9, _ => num = 19, num1 = 10, n1um = 100
#     2. Not allowed characters - Special chracters => Num$ = 100 => Not acceptable
#     3. Identifier should not start with the number => 1num = 100 => Not acceptable
#     4. Do not use reserved words as an identifier
#     5. Not recommended to take too lengthy identifier
#     6. _abc - Private - you can not access it outside the module
#        __abc - Strongly private - can not access it outside the class in the same module
#        __abc__ - Language defined special identifier - __add__, __name__, __doc__
# print(10+20)
# __add__ = 100
# print(10+20)
```
